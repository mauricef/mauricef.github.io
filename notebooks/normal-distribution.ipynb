```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import comb
```

## Normal Distribution
We perform an experiment which consists of flipping a fair coin N times, we count the number of heads. We perform
K trials of this experiment. At the end we have a list of K integers, each integer is between 0 and N (the number of heads for each trial)

```python
N = 20
K = 100000
results = np.sum(np.random.choice([0, 1], size=(K, N)), axis=1)
results[0:100]
```

Now we count up the frequencies of each result and plot it as a density bar chart

```python
frequencies = Counter(results)
x = np.arange(0, N+1)
y = np.array([frequencies[xi]/K for xi in x])
plt.bar(x, y)
plt.show()
```

This distribution is approximately normal, with mean of N/2 and variance of sqrt(N)

```python
def normal(x, mean, var):
    return (1 / np.sqrt(2 * np.pi * var)) * np.exp(-((x - mean)**2) / (2 * var))
```

```python
xn = np.arange(0, N+1, .1)
yn = normal(xn, mean=N/2., var=np.sqrt(N))
plt.bar(x, y, alpha=.5)
plt.plot(xn, yn)
plt.show()
```
