```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
```

# Frequency analysis using Scipy
This notebook demonstrates how to use `fftpack.fft` to analyze the spectrum of a simple 1d timeseries.

## Generate the signal
We generate a timeseries as a mixture of two sine waves.

```python
N = 24 * 7 * 4
DAILY = 24
WEEKLY = 24 * 7
x = np.arange(0, N)
decay = np.exp(-.005*x)
daily = np.sin(2.0 * np.pi * x / DAILY)
weekly = np.sin(2.0 * np.pi * x / WEEKLY)
y = (daily + weekly) * decay
```

```python
plt.plot(x, y)
plt.show()
```

## Infer the frequencies
We use `fftpack.fft` to recover the frequencies from the discrete timeseries.

```python
yf = scipy.fftpack.fft(y)
yf = yf[:len(yf)//2]
yf = yf[:100]
yf = np.abs(yf)
plt.plot(yf)
plt.show()
```

```python
peaks = np.argsort(yf)[::-1][:2]
```

```python
set(N/peaks) == set([DAILY, WEEKLY])
```
